```python
import pandas as pd
import numpy as np
import tqdm as tqdm
import os
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from sklearn.utils import shuffle
from sklearn.metrics import classification_report,confusion_matrix
import torch
device = torch.device("mps")
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, TensorBoard, ModelCheckpoint
```

```python
labels = ['glioma_tumor','no_tumor','meningioma_tumor','pituitary_tumor']
```

```python
X_train = []
y_train = []
image_size = 150
for i in labels:
    folderPath = os.path.join('/Users/pranavmoses/Desktop/5505/project/tumor.v3i.folder','train',i)
    for j in os.listdir(folderPath):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size, image_size))
        X_train.append(img)
        y_train.append(i)
        
X_test = []
y_test = []
for i in labels:
    folderPath = os.path.join('/Users/pranavmoses/Desktop/5505/project/tumor.v3i.folder','test',i)
    for j in os.listdir(folderPath):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size,image_size))
        X_test.append(img)
        y_test.append(i)
        
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
np.unique(y_train)
```

```python
plt.imshow(X_train[0])
```

```python
y_train = np.array(pd.get_dummies(y_train, dtype=float))
y_test = np.array(pd.get_dummies(y_test, dtype=float))
```

```python
X_train, y_train = shuffle(X_train,y_train, random_state=101)
```

```python
X_test, y_test = shuffle(X_test,y_test, random_state=101)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
effnet = EfficientNetB0(weights='imagenet',include_top=False,input_shape=(image_size,image_size,3))
```

```python
effnet.summary()
```

```python
import tensorflow as tf
```

```python
model = effnet.output
model = tf.keras.layers.GlobalAveragePooling2D()(model)
model = tf.keras.layers.Dropout(rate=0.5)(model)
model = tf.keras.layers.Dense(4,activation='softmax')(model)
model = tf.keras.models.Model(inputs=effnet.input, outputs = model)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer = 'Adam', metrics= ['categorical_accuracy'])
```

```python
tf.config.run_functions_eagerly(True)
```

```python
tensorboard = TensorBoard(log_dir = '/Users/pranavmoses/Desktop/5505/project/logs')
```

```python
checkpoint = ModelCheckpoint("effnet.keras", monitor="val_accuracy", save_best_only=True, mode="auto", verbose=1)
```

```python
reduce_lr = ReduceLROnPlateau(monitor = 'val_accuracy', factor = 0.3, patience = 2, min_delta = 0.001, mode='auto',verbose=1)
```

```python
model.fit(X_train,y_train,validation_split=0.1, epochs =4, verbose=1, batch_size=32)
```

```python
model
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
pred = np.argmax(pred,axis=1)
```

```python
np.unique(y_test)
```

```python
y_test_pred = pd.get_dummies(pred, dtype=float)
```

```python
np.unique(y_test_pred)
```

```python
print(classification_report(y_test,y_test_pred))
```
